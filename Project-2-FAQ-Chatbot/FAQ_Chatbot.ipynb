```python
import nltk
import string
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
nltk.download('punkt_tab')
nltk.download('stopwords')
faqs = {
    "What is the admission process?":
    "The admission process includes filling out an application form and attending counseling.",

    "What documents are required for admission?":
    "You need mark sheets, ID proof, passport size photos, and transfer certificate.",

    "What is the fee structure?":
    "The annual fee structure depends on the course selected.",

    "Is hostel facility available?":
    "Yes, hostel facilities are available for both boys and girls.",

    "What courses are offered?":
    "We offer B.Tech, M.Tech, MBA, and MCA programs."
}
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)
questions = list(faqs.keys())
answers = list(faqs.values())
processed_questions = [preprocess(q) for q in questions]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_questions)
def chatbot_response(user_input):
    user_input_processed = preprocess(user_input)
    user_vector = vectorizer.transform([user_input_processed])

    similarity = cosine_similarity(user_vector, X)
    index = np.argmax(similarity)

    return answers[index]
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Thank you!")
        break
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```
